```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#x = np.linspace(1,2,20)
#y = np.linspace(2,3,20)
x = np.arange(1,2,0.05)
y = np.arange(2,3,0.05)

plt.figure(num=0,figsize=[15,5])
plt.subplot(131)
plt.plot(x,y,marker="x")
plt.title('Linear line')
#plt.show()


y = y + 1/2*np.random.rand(np.size(y))

#plt.figure(num=1,figsize=[5,5])
plt.subplot(132)
plt.plot(x,y,marker="x")
plt.title('Randomized')
plt.show()

w = 1/4*np.random.rand(1)
b = 1/4*np.random.rand(1)

alpha = 0.001

N = np.size(y)

input_ = x

for i in range(0,50):
    for k in range(0,N):
        output = np.dot(np.asarray([w,b]).T,np.asarray([input_[k],1]))
    
        dedw = input_[k]
    
        dLde = output - y[k]
    
        dedb = 1
    
        dLdw = dLde*dedw
    
        dLdb = dLde*dedb
    
        w = w - alpha*dLdw
    
        b = b - alpha*dLdb
    
#    plt.subplot(133)
    plt.plot(x,w*x+b,'-r', x,y,marker="x")
    plt.title('Regression by Gradient Descent method')
    plt.show()

print("weight: ", w)
print("bias: ", b)




return



```
